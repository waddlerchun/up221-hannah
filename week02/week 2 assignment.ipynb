```python
# import data: read and add .csv file to notebook
svi = pd.read_csv('data/Los_Angeles_County_CVA_Social_Vulnerability_Index.csv')         
```

There are 2327 rows and 46 columns in the dataset.

```python
# dataframe column information
svi.info()
```

There are 29 indicators for social vulnerability.The 41st column holds the information on the SVI score. 

```python
# dataframe row information
svi.head()
```

Each row correlates to a location with an assigned SVI score. 

```python
# How many times is a city represented in the LA County Census Tract?
# Store this data in the variable city_count
city_count = svi['CSA_Label'].value_counts()
city_count
```

The City of Long Beach is represented in the Census Tract the most - 108 times.

```python
# import graphical plotting library
import matplotlib.pyplot as plt
```

```python
city_count.plot.bar()
```

This plot is messy. The data will be trimmed to only include the Incorporated San Fernando Valley Planning Area.

```python
# filter the dataset to only include the Santa Monica Mountains Planning Area and save
svi_trimmed = svi[svi["DRP_Planning_Area"] == 'Santa Monica Mountains Planning Area'].copy()
svi_trimmed
```

There are only 19 rows, compared to the original 2327 rows.

Let's try plotting again.

```python
# Plot the CSA Label against the SVI 
svi_trimmed.plot.bar (x='CSA_Label', y = 'SoVI_Score')
```

This bar chart indicates that all the Census Tract locations within the Santa Monica Mountains Planning Area have negative scores on the social vulnerability index. This means this Distribution Resource Planning (DRP) Area is rated as not being a vulnerable area to climate change.
